```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Country 1

```python
data = pd.read_excel("Country1.xlsx")
```

```python
data
```

```python
forecast_start = 154  # The index where forecast starts

# Plot historical and future data separately for each column
columns = ['region_1', 'region_2', 'region_3', 'national']

# Set up the figure and axes
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 15))
axes = axes.flatten()

# Loop over each column (region_1, region_2, region_3, national)
for i, col in enumerate(columns):
    # Plot historical data
    axes[i].plot(data['date'][:forecast_start], data[col][:forecast_start], label='Historical', color='blue')
    
    # Plot forecasted data
    axes[i].plot(data['date'][forecast_start:], data[col][forecast_start:], label='Forecast', color='red')
    
    # Add labels and title
    axes[i].set_title(f"{col.capitalize()} - Historical vs Forecasted")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Sales')
    axes[i].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
